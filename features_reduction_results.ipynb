```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("feature_reduction_results.csv")
df
```

```python
problems = df.PROBLEM.unique()
problems
```

```python
feature_sets = df[df.FEATURE_SET != "-"].FEATURE_SET.unique()
feature_sets
```

```python
resutls_accuracy = pd.DataFrame()
resutls_accuracy["FEATURE_SET"] = pd.Series(feature_sets)
for problem in problems:
    resutls_accuracy[problem] =  (df[df.PROBLEM == problem][df.FEATURE_SET != "-"].ACCURACY - list(df[df.PROBLEM == problem][df.FEATURE_SET == "-"].ACCURACY)[0]).reset_index().ACCURACY
resutls_accuracy
```

```python
resutls_accuracy["MEAN"] = resutls_accuracy.mean(axis=1)
resutls_accuracy["STD"] = resutls_accuracy.std(axis=1)
resutls_accuracy[["FEATURE_SET", "MEAN", "STD"]].sort_values("MEAN")
```
